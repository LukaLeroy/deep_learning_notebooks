```python
temperature = df['T (degC)'].values
raw_data = df.drop(['Date Time'], axis=1).values
```

```python
raw_data[:3]
```

```python
raw_data.shape
```

```python
num_train_samples = raw_data.shape[0] // 2
num_val_samples = raw_data.shape[0] // 4
```

```python
mean = raw_data[:num_train_samples].mean(axis=0)
std = raw_data[:num_train_samples].std(axis=0)
```

```python
raw_data = (raw_data - mean) /std
```

```python
raw_data[:num_train_samples].std(axis=0)
```

```python
def create_ds(raw_data, target, shuffle=True, batch_size=256):

  delay = 120 * 6 + (24 -1) * 6
  return tf.keras.utils.timeseries_dataset_from_array(
      data=raw_data,
      targets=target[delay:],
      sequence_length=120,
      sampling_rate=6,
      batch_size=batch_size,
      shuffle=shuffle
  )
```

```python
my_check_ds = create_ds(
  raw_data=df.values[:1000,0],
  target=df.values[:1000,0],
  shuffle=False,
  batch_size=1
)
for features, target in my_check_ds.take(2):
  print(f"{features} maps to {target}")
  print("*"*50)
```

```python
train_ds = create_ds(
    raw_data[:num_train_samples],
    temperature[:num_train_samples],
    shuffle=True
)
```

```python
val_ds = create_ds(
    raw_data[num_train_samples:num_train_samples+num_val_samples],
    temperature[num_train_samples:num_train_samples+num_val_samples],
    shuffle=False
)
```

```python
test_ds = create_ds(
    raw_data[num_train_samples+num_val_samples:],
    temperature[num_train_samples+num_val_samples:],
    shuffle=False
)
```

```python
for samples, targets in train_ds.take(1):
  print(samples.shape)
  print(targets.shape)
```

```python
def compute_naive_mae(dataset):
  mean = temperature[:num_train_samples].mean()
  std = temperature[:num_train_samples].std()

  total_error = 0.0
  total_samples = 0
  for samples, target in dataset:
    current_temperature = samples[:, -1, 1] * std + mean
    total_error += np.sum(np.abs(current_temperature - target))
    total_samples += samples.shape[0]

  return total_error / total_samples
```

```python
compute_naive_mae(val_ds)
```

```python
compute_naive_mae(test_ds)
```

```python
train_ds = train_ds.cache().prefetch(1)
val_ds = val_ds.cache().prefetch(1)
```

```python
def get_model():
  model = tf.keras.Sequential()
  model.add(tf.keras.layers.Flatten(input_shape=(120, 14)))
  model.add(tf.keras.layers.Dense(16, activation='relu', ))
  model.add(tf.keras.layers.Dense(1, activation=None))

  return model
```

```python
class MyLSTMCell(tf.keras.layers.Layer):
  def __init__(self, units, **kwargs):
    super().__init__(**kwargs)
    self.state_size = [units, units]
    self.output_size = units

    self.dense_i = tf.keras.layers.Dense(
        units=units,
        bias_initializer="zeros",
        activation="sigmoid")

    self.dense_f = tf.keras.layers.Dense(
        units=units,
        bias_initializer="ones",
        activation="sigmoid"
    )

    self.dense_g = tf.keras.layers.Dense(
        units=units,
        bias_initializer="zeros",
        activation="tanh"
    )

    self.dense_o = tf.keras.layers.Dense(
        units=units,
        bias_initializer="zeros",
        activation="sigmoid"
    )

  def call(self, inputs, states):
    previous_short_term_state = states[0]
    previous_long_term_state = states[1]

    input_and_previous_short_term = tf.concat([inputs,previous_short_term_state], axis=-1)

    i = self.dense_i(input_and_previous_short_term)
    f = self.dense_f(input_and_previous_short_term)
    g = self.dense_g(input_and_previous_short_term)
    o = self.dense_o(input_and_previous_short_term)

    new_carry_state = f * previous_short_term_state + i * g
    y = o * tf.keras.activations.tanh(new_carry_state)

    return y, [y, new_carry_state]
```

```python
BATCH_SIZE = 5
NUM_STEPS = 10
NUM_FEATURES = 3
UNITS=4
inputs = tf.random.normal([BATCH_SIZE, NUM_STEPS, NUM_FEATURES])
```

```python
lstm = MyLSTMCell(units=UNITS)

model = tf.keras.Sequential([
    tf.keras.layers.RNN(lstm)
])
```

```python
output = model(inputs)
```
